### Distribution of Class Instances per Recording
- average number of passengers, maximum number of passengers
- For each category_id count unique tracking_ids for annotation

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from pathlib import Path
import pandas as pd
import json

from src import DATA_BASE_DIR as data_base_dir
from src.data import list_record_ids
record_ids = list_record_ids()
from src.visualization import VIZ_PARAMS
bbox_colors = VIZ_PARAMS['gt_bbox_colors']
```

```python
records_data = []

for i, record_id in enumerate(record_ids):
    annotation_path = Path(data_base_dir) / "annotations" / f"{record_id}_annotations.json"
    if not annotation_path.is_file():
        raise FileNotFoundError(f"Json file not found: {annotation_path}")
    
    with open(annotation_path, 'r') as f:
        all_annotations = json.load(f)
    
    category_id_name_map = {category['id']: category['name'] for category in all_annotations.get("categories", [])}
    
    class_counts = {
    name: len({ann["attributes"]["track_id"]
               for ann in all_annotations.get("annotations", [])
               if ann["category_id"] == cid})
    for cid, name in category_id_name_map.items()
    }

    records_data.append({"record_id": record_id, **class_counts})

df = pd.DataFrame(records_data, columns=records_data[0].keys())
df = df.replace(0, np.nan)
```

```python
df
```

```python
xmin= -1
xmax = len(record_ids)-0.4
ymin = 0
ymax = 7.2

x_sep0 = (record_ids.index("rec0") + record_ids.index("rec1")) / 2
x_sep = (record_ids.index("rec25") + record_ids.index("rec26")) / 2

plt.figure(figsize=(2.6, 2.2), dpi=200)
plt.bar(
    record_ids,
    df.person,
    color=bbox_colors['person'],
    label='Person'
)


plt.bar(
    record_ids,
    df.clothing,
    bottom=df.person.fillna(0),
    color=bbox_colors['clothing'],
    label='Clothing'
)

plt.bar(
    record_ids,
    df.bag,
    bottom=(df.person.fillna(0) +
            df.clothing.fillna(0)),
    color=bbox_colors['bag'],
    label='Bag'
)

#plt.hlines(xmin=x_sep0, xmax=xmax, y=df.person.mean(), linestyle='-', color='black', label=f'Mean (Person): {df.person.mean():.1f}')


plt.axvline(x=x_sep, color="black", linestyle="--", linewidth=1)
plt.axvline(x=x_sep0, color="black", linestyle="--", linewidth=1)

# --- Labels for Training and Validation ---
plt.text(
    (x_sep0 + x_sep) / 2,                     # midpoint of left section
    ymax*0.92,                      
    "Training",
    ha='center', va='center',
    fontweight='bold'
)

plt.text(
    (x_sep + xmax) / 2,            # midpoint of right section
    ymax*0.92,                   
    "Val.",
    ha='center', va='center',
    fontweight='bold'
)

plt.text(0, ymax * 0.22, "Background only", rotation='vertical', ha='center', va='center', fontsize=7)


plt.ylim(ymin, ymax)
plt.xlim(xmin,xmax)
plt.legend(loc='upper left', framealpha=1, frameon=True)
plt.xlabel('Record ID')
plt.xticks(rotation=90)
plt.ylabel('Distinct Track IDs per Class')
plt.tight_layout(pad=0.05)
plt.show()
```

## Distribution Action per Bounding Box per Recording

```python
import json
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np

from src import DATA_BASE_DIR as data_base_dir
from src.data import list_record_ids
record_ids = list_record_ids()
from src.visualization import VIZ_PARAMS
bbox_colors = VIZ_PARAMS['gt_bbox_colors']
```

```python
records_action_data = []
total_person_count = 0

for record_id in record_ids:
    annotation_path = Path(data_base_dir) / "annotations" / f"{record_id}_annotations.json"
    if not annotation_path.is_file():
        raise FileNotFoundError(f"Json file not found: {annotation_path}")
    
    with open(annotation_path, 'r') as f:
        all_annotations = json.load(f)
    
    # Filter to only person annotations (or adjust if actions exist for other classes too)
    person_anns = [
        ann for ann in all_annotations.get("annotations", [])
        if all_annotations["categories"][ann["category_id"] - 1]["name"] == "person"
    ]
    person_count = len(person_anns)
    total_person_count += person_count
    
    # Extract actions
    actions = [ann["attributes"].get("action") for ann in person_anns if "action" in ann["attributes"]]
    
    # Count frequency of each action
    action_counts = dict(Counter(actions))
    
    # Add record_id to the row
    records_action_data.append({"record_id": record_id, **action_counts})

# Create DataFrame
df_act = pd.DataFrame(records_action_data).fillna(0)
df_act = df_act.replace(0, np.nan)
print(total_person_count)
```

```python
xmin= -1
xmax = len(record_ids)-0.4
ymin = 0
ymax = 3500

x_sep0 = (record_ids.index("rec0") + record_ids.index("rec1")) / 2
x_sep = (record_ids.index("rec25") + record_ids.index("rec26")) / 2

plt.figure(figsize=(2.6, 2.2), dpi=200)

b_seated         = df_act.seated
b_seated_ground  = df_act.seated.fillna(0)
b_standing       = (df_act.seated.fillna(0) +
                    df_act.seated_ground.fillna(0))
b_lying          = (df_act.seated.fillna(0) +
                    df_act.seated_ground.fillna(0) +
                    df_act.standing.fillna(0))


plt.bar(record_ids, df_act.seated,
        color=bbox_colors['seated'], label='seated',
        edgecolor='none', antialiased=False)

plt.bar(record_ids, df_act.seated_ground, bottom=b_seated_ground,
        color=bbox_colors['seated_ground'], label='seated (ground)',
        edgecolor='none', antialiased=False)

plt.bar(record_ids, df_act.standing, bottom=b_standing,
        color=bbox_colors['standing'], label='standing',
        edgecolor='none', antialiased=False)

plt.bar(record_ids, df_act.lying, bottom=b_lying,
        color=bbox_colors['lying'], label='lying',
        edgecolor='none', antialiased=False)


plt.axvline(x=x_sep0, color="black", linestyle="--", linewidth=1)
plt.axvline(x=x_sep, color="black", linestyle="--", linewidth=1)
# --- Labels for Training and Validation ---
plt.text(
    (x_sep / 2),                     # midpoint of left section
    ymax*0.92,                      
    "Train.",
    ha='center', va='center', fontweight='bold'
)

plt.text(
    (x_sep + xmax) / 2,            # midpoint of right section
    ymax*0.5,                   
    "Val.",
    ha='center', va='center',
    fontweight='bold'
)

plt.text(0, ymax * 0.22, "Background only", rotation='vertical', ha='center', va='center', fontsize=7)


plt.ylim(ymin, ymax)
plt.xlim(xmin,xmax)
plt.legend(loc='upper right',framealpha=1, frameon=True)
plt.xlabel('Record ID')
plt.xticks(rotation=90)
plt.ylabel('Count of Action Instances')
plt.tight_layout(pad=0.05)
plt.show()

```
